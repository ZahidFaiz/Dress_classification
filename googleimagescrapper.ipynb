```
def search_and_download(search_term: str, driver_path = '/content/chromedriver', target_path='/content/downloads', number_images= 300):
    target_folder = os.path.join(target_path, '_'.join(search_term.lower().split(' ')))

    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    with webdriver.Chrome('chromedriver',chrome_options=chrome_options) as wd:
        res = fetch_image_urls(search_term, number_images, wd=wd, sleep_between_interactions=2.5)
        
    for elem in res:
        persist_image(target_folder, elem)
        
```

```
import os
import hashlib
import io
import os
from time import sleep
import requests
from pip._vendor import *
import pip
from PIL import Image as Image
from selenium import webdriver
from selenium import webdriver
# from webdriver_manager.chrome import ChromeDriverManager

# driver = webdriver.Chrome(ChromeDriverManager().install())


def fetch_image_urls(query: str, max_links_to_fetch: int , wd :  webdriver, sleep_between_interactions= 1):
    def scroll_to_end(wd):
        lenOfPage = wd.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        #time.sleep(sleep_between_interactions)
        sleep(2)
        match = False
        while (match == False):
            lastCount = lenOfPage
            sleep(5)
            lenOfPage = wd.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
            if lastCount == lenOfPage:
                match = True
        # build the google query

    # search_url = "https://www.google.com/search?safe=off&site=&tbm=isch&source=hp&q={q}&oq={q}&gs_l=img"
    # search_url = "https://www.google.com/search?q=saree%20cotton%20pinterest%20customer&tbm=isch&tbs=rimg%3ACUtojuOFtXnJImDZeN_1DNMIAwTBcas4EPrJiYWJHcA1CfGcW7xkOg_12KAW1AIEO3zUe10eCUSVi2SkHSTgwPtuTx3TL9Bd_1_1B3JXGxWnixML3maPU8N_125VDc9rFp5WK7UpNw-TFOEZluSAqEgnZeN_1DNMIAwRHHJl7nfM5zNyoSCTBcas4EPrJiERa2MD0tnHUzKhIJYWJHcA1CfGcRwS1MgFSJnAUqEgkW7xkOg_12KARHGW5KNRmOQzyoSCW1AIEO3zUe1EYny1m-LvgQwKhIJ0eCUSVi2SkERMQMwrbuP7Z0qEgnSTgwPtuTx3RFHPz81PcNTkSoSCTL9Bd_1_1B3JXET7AdaihZMHCKhIJGxWnixML3mYREqyUe2h0kzUqEgmPU8N_125VDcxFENVH_1Fxn5gCoSCdrFp5WK7UpNER_1bTsRZr5zxKhIJw-TFOEZluSARCyaj5leW4BNhs22I7OuFwck&hl=en&ved=0CBwQuIIBahcKEwjg24qkpKbpAhUAAAAAHQAAAAAQBg&biw=1534&bih=793"
    # search_url = "https://www.google.com/search?q=saree+amazon&tbm=isch&ved=2ahUKEwj7r7ryzKbpAhW-03MBHfedBygQ2-cCegQIABAA"
    # search_url = "https://www.google.com/search?q=saree+pinterest+plain&tbm=isch&ved=2ahUKEwjSq-LTy6bpAhUzMbcAHTLwBugQ2-cCegQIABAA&oq=saree+pinterest+plain&gs_lcp=CgNpbWcQAzIGCAAQCBAeMgYIABAIEB4yBggAEAgQHjIGCAAQCBAeOgQIIxAnOgIIADoECAAQHjoGCAAQBRAeUIIuWIw2YKg4aABwAHgAgAHXAYgBuQmSAQUwLjIuNJgBAKABAaoBC2d3cy13aXotaW1n&sclient=img&ei=YIi2XtL7NrPi3LUPsuCbwA4&bih=793&biw=1544&client=ubuntu&hs=ua5"
    # search_url = "https://www.google.com/search?q=saree+south+indian&tbm=isch&ved=2ahUKEwizuKDYz6bpAhXJ03MBHW_rDtoQ2-cCegQIABAA&oq=saree+south&gs_lcp=CgNpbWcQARgCMgIIADICCAAyAggAMgIIADICCAAyAggAMgIIADICCAAyAggAMgIIADoECCMQJzoFCAAQgwFQkSFY-iZgvz1oAHAAeACAAZsDiAGsDpIBBTMtMy4ymAEAoAEBqgELZ3dzLXdpei1pbWc&sclient=img&ei=nIy2XrP6Asmnz7sP79a70A0&bih=793&biw=1534&hl=en"
    # search_url = "https://www.google.com/search?q=saree+south+indian&tbm=isch&hl=en&chips=q:saree+south+indian,g_1:kerala&hl=en&ved=2ahUKEwiL_Ibez6bpAhUfJbcAHQnnDvMQ4lYoBHoECAEQGA&biw=1534&bih=793"
    # search_url = "https://www.google.com/search?q=saree+south+indian&tbm=isch&ved=2ahUKEwizuKDYz6bpAhXJ03MBHW_rDtoQ2-cCegQIABAA&oq=saree+south&gs_lcp=CgNpbWcQARgCMgIIADICCAAyAggAMgIIADICCAAyAggAMgIIADICCAAyAggAMgIIADoECCMQJzoFCAAQgwFQkSFY-iZgvz1oAHAAeACAAZsDiAGsDpIBBTMtMy4ymAEAoAEBqgELZ3dzLXdpei1pbWc&sclient=img&ei=nIy2XrP6Asmnz7sP79a70A0&bih=793&biw=1534&hl=en"
        # load the page
    links_saree = [
                    "https://www.google.com/search?q=saree+south+indian&tbm=isch&hl=en&chips=q:saree+south+indian,g_1:kerala&hl=en&ved=2ahUKEwiL_Ibez6bpAhUfJbcAHQnnDvMQ4lYoBHoECAEQGA&biw=1534&bih=793",
                    "https://www.google.com/search?q=saree+south+indian&tbm=isch&hl=en&chips=q:saree+south+indian,g_1:tamil&hl=en&ved=2ahUKEwiL_Ibez6bpAhUfJbcAHQnnDvMQ4lYoCHoECAEQHA&biw=1534&bih=793",
                    "https://www.google.com/search?q=saree+full+sleeve+&tbm=isch&ved=2ahUKEwiAm6u50KbpAhWGGrcAHXXrAy8Q2-cCegQIABAA&oq=saree+full+sleeve+&gs_lcp=CgNpbWcQA1DRRFjRRGDGRmgAcAB4AIABAIgBAJIBAJgBAKABAaoBC2d3cy13aXotaW1n&sclient=img&ei=Z422XsDLJ4a13LUP9daP-AI",
                    "https://www.google.com/search?q=saree+pinterest+plain&tbm=isch&ved=2ahUKEwi4tJ3l0KbpAhV1DbcAHYXiAJcQ2-cCegQIABAA&oq=saree+pintere&gs_lcp=CgNpbWcQARgAMgQIIxAnMgQIIxAnMgIIADICCAAyBggAEAUQHjIGCAAQBRAeMgYIABAFEB4yBggAEAUQHjIGCAAQBRAeMgYIABAFEB46BAgAEENQ2a0oWIbCKGCdzShoAHAAeACAAdYEiAGNGJIBCzAuNC4xLjMuMS4xmAEAoAEBqgELZ3dzLXdpei1pbWc&sclient=img&ei=w422XvjGKvWa3LUPhcWDuAk&bih=793&biw=1534&hl=en",
                    "https://www.google.com/search?q=saree+pinterest+white&tbm=isch&ved=2ahUKEwjh-faj06bpAhXK2HMBHSJmBZgQ2-cCegQIABAA&oq=saree+pinterest+white&gs_lcp=CgNpbWcQAzIGCAAQCBAeUMTsClix8Qpg7fIKaABwAHgAgAGlBIgBkAqSAQkyLTEuMC4xLjGYAQCgAQGqAQtnd3Mtd2l6LWltZw&sclient=img&ei=YJC2XuHJA8qxz7sPosyVwAk&bih=793&biw=1534&hl=en",
                    "https://www.google.com/search?q=saree+pinterest+uniform+&tbm=isch&ved=2ahUKEwikpsis1KbpAhWvMrcAHTxzAbAQ2-cCegQIABAA&oq=saree+pinterest+uniform+&gs_lcp=CgNpbWcQAzoECCMQJzoCCAA6BAgAEB46BggAEAUQHlDxKViSSGC1SWgBcAB4AIABoQSIAYIUkgELMC42LjEuMS4xLjGYAQCgAQGqAQtnd3Mtd2l6LWltZw&sclient=img&ei=fpG2XqTvJK_l3LUPvOaFgAs&bih=793&biw=1534&hl=en",
                    "https://www.google.com/search?q=saree+pinterest+uniform+&tbm=isch&ved=2ahUKEwikpsis1KbpAhWvMrcAHTxzAbAQ2-cCegQIABAA&oq=saree+pinterest+uniform+&gs_lcp=CgNpbWcQAzoECCMQJzoCCAA6BAgAEB46BggAEAUQHlDxKViSSGC1SWgBcAB4AIABoQSIAYIUkgELMC42LjEuMS4xLjGYAQCgAQGqAQtnd3Mtd2l6LWltZw&sclient=img&ei=fpG2XqTvJK_l3LUPvOaFgAs&bih=793&biw=1534&hl=en"
                    ]



    image_urls = set()
    image_count = 0
    results_start = 0
    
    for search_url in links_saree:
      wd.get(search_url.format(q=query))

      while image_count < max_links_to_fetch:
          sleep(6)
          scroll_to_end(wd)

          # get all image thumbnail results
          thumbnail_results = wd.find_elements_by_css_selector("img.Q4LuWd")
          number_results = len(thumbnail_results)

          print(f"Found: {number_results} search results. Extracting links from {results_start}:{number_results}")

          for img in thumbnail_results[results_start:number_results]:
              # try to click every thumbnail such that we can get the real image behind it
              try:
                  img.click()
                  sleep(5)
              except Exception:
                  continue

              # extract image urls
              actual_images = wd.find_elements_by_css_selector('img.n3VNCb')
              for actual_image in actual_images:
                  if actual_image.get_attribute('src') and 'http' in actual_image.get_attribute('src'):
                      image_urls.add(actual_image.get_attribute('src'))

              image_count = len(image_urls)

              if len(image_urls) >= max_links_to_fetch:
                  print(f"Found: {len(image_urls)} image links, done!")
                  break
          else:
              print("Found:", len(image_urls), "image links, looking for more ...")
              sleep(20)
              return
              load_more_button = wd.find_element_by_css_selector(".mye4qd")
              if load_more_button:
                  wd.execute_script("document.querySelector('.mye4qd').click();")

          # move the result startpoint further down
          results_start = len(thumbnail_results)

    return image_urls

def persist_image(folder_path:str,url:str):
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - Could not download {url} - {e}")

    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        file_path = os.path.join(folder_path,hashlib.sha1(image_content).hexdigest()[:10] + '.jpg')
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
        print(f"SUCCESS - saved {url} - as {file_path}")
    except Exception as e:
        print(f"ERROR - Could not save {url} - {e}")
```
